```python
# =============================================================
#  G1_UsoPandas.py
#  Evaluación Sumativa – Introducción a las funcionalidades de Pandas
#  Fuente de datos: TablaPosiciones2025.xlsx
# =============================================================

import pandas as pd

# ══════════════════════════════════════════════════════════════
# 1. DEFINICIÓN DE DATAFRAME
#    Cargamos el archivo Excel y generamos el DataFrame inicial.
# ══════════════════════════════════════════════════════════════

print("=" * 60)
print("1. DEFINICIÓN DE DATAFRAME")
print("=" * 60)

# leer de acuerdo al nombre del archivo
df = pd.read_csv("TablaPosiciones2025_G10.csv", encoding="cp1252")

# Renombramos las columnas para mayor claridad
df.columns = [
    "Posicion", "Equipo", "PT", "PJ", "PG", "PP", "PE",
    "GF", "GC", "GFV", "DG"
]

print("\nDataFrame inicial (primeras 5 filas):")
print(df.head())
print(f"\nDimensiones del DataFrame: {df.shape[0]} filas x {df.shape[1]} columnas")


# ══════════════════════════════════════════════════════════════
# 2. CONFIGURACIÓN DE ÍNDICES – 4 SERIES
#    • 2 Series a partir de COLUMNAS del DataFrame
#    • 2 Series a partir de FILAS del DataFrame
# ══════════════════════════════════════════════════════════════

print("\n" + "=" * 60)
print("2. CONFIGURACIÓN DE ÍNDICES – SERIES")
print("=" * 60)

# Usamos la columna 'Equipo' como índice para que las series
# sean más fáciles de leer
df_idx = df.set_index("Equipo")

# --- Serie 1: Puntos (PT) de cada equipo ---
serie_puntos = df_idx["PT"]
print("\nSerie 1 – Puntos (PT) por equipo:")
print(serie_puntos.to_string())

# --- Serie 2: Diferencia de goles (DG) de cada equipo ---
serie_dg = df_idx["DG"]
print("\nSerie 2 – Diferencia de goles (DG) por equipo:")
print(serie_dg.to_string())

# --- Serie 3: Todos los datos del equipo en la fila 0 (Independiente Medellín) ---
serie_medellin = df.loc[0]
print("\nSerie 3 – Datos de Independiente Medellín (fila 0):")
print(serie_medellin.to_string())

# --- Serie 4: Todos los datos del equipo en la fila 8 (Junior) ---
serie_junior = df.loc[8]
print("\nSerie 4 – Datos de Junior (fila 8):")
print(serie_junior.to_string())


# ══════════════════════════════════════════════════════════════
# 3. EXPLORACIÓN DE DATOS
#    Tres funciones para explorar y resumir el DataFrame.
# ══════════════════════════════════════════════════════════════

print("\n" + "=" * 60)
print("3. EXPLORACIÓN DE DATOS")
print("=" * 60)

# --- Función 1: describe() – estadísticas descriptivas ---
print("\nFunción 1 – describe(): Estadísticas descriptivas de las columnas numéricas")
print(df.describe().round(2).to_string())

# --- Función 2: info() – estructura del DataFrame ---
print("\nFunción 2 – info(): Estructura y tipos de datos del DataFrame")
df.info()

# --- Función 3: value_counts() – distribución de victorias ---
print("\nFunción 3 – value_counts(): Distribución de partidos ganados (PG)")
print(df["PG"].value_counts().sort_index().to_string())
print("  → Muestra cuántos equipos tienen 0, 1, 2, 3 o 4 victorias")


# ══════════════════════════════════════════════════════════════
# 4. FILTRADO DE DATOS
#    • iloc  – selección por posición numérica
#    • Dos condicionales (filtros lógicos)
# ══════════════════════════════════════════════════════════════

print("\n" + "=" * 60)
print("4. FILTRADO DE DATOS")
print("=" * 60)

# --- iloc: primeros 5 equipos, columnas Equipo / PT / PJ / DG ---
print("\nilic – Top 5 equipos (filas 0-4), columnas seleccionadas:")
print(df.iloc[0:5, [1, 2, 3, 10]].to_string(index=False))

# --- Condicional 1: equipos con más de 10 puntos ---
filtro1 = df[df["PT"] > 10]
print("\nCondicional 1 – Equipos con más de 10 puntos (PT > 10):")
print(filtro1[["Posicion", "Equipo", "PT", "PG", "DG"]].to_string(index=False))

# --- Condicional 2: equipos con diferencia de goles positiva Y
#     que hayan jugado al menos 6 partidos ---
filtro2 = df[(df["DG"] > 0) & (df["PJ"] >= 6)]
print("\nCondicional 2 – Equipos con DG positiva Y PJ ≥ 6:")
print(filtro2[["Posicion", "Equipo", "PJ", "GF", "GC", "DG"]].to_string(index=False))

print("\n" + "=" * 60)
print("Programa finalizado correctamente.")
print("=" * 60)
```
