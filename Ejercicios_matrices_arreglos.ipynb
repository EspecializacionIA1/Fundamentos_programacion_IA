```python
import numpy as np
A = np.array([[1, 2], [3, 4]])
B = A @ A - 2 * A
print(np.linalg.det(B))
```

```python
import numpy as np
v = np.array([1, 2, 2])
v_norm = np.linalg.norm(v)
P = np.outer(v, v) / (v_norm**2)
print(np.round(P @ v - v, decimals=10))
```

```python
import numpy as np
U = np.random.randn(5, 3)
A = U @ U.T
print(np.linalg.matrix_rank(A))
```

```python
import numpy as np
A = np.array([[0, 1], [-1, 0]])
v = np.array([1, 0])
for _ in range(4):
    v = A @ v
print(v)
```

```python
import numpy as np
A = np.array([[1, 1], [0, 1]])
logdet = np.log(np.linalg.det(A @ A.T))
print(np.round(logdet, 6))
```
