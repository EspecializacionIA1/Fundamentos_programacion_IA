```python
# ----------------------------------------------------------------------------------------
# PROGRAMA: Solución de ecuación cuadrática
# ----------------------------------------------------------------------------------------
# Descripción:
# Este programa calcula las soluciones de una ecuación cuadrática
# de la forma ax^2 + bx + c = 0.
# Para este programa se utilizó la creación de variable para calcular el
# determinante, con el cual podremos saber si la ecuación no tiene solución,
# más de una solución? tiene dos soluciones?
# ----------------------------------------------------------------------------------------
# Autor: Juan Ayala
# Version: 1.0
# [01.02.2026]
# ----------------------------------------------------------------------------------------

# ----------------------------------------------------------------------------------------
# IMPORTAR MODULOS
# ----------------------------------------------------------------------------------------
import math  # modulo para operaciones matemáticas (raíz cuadrada)

# ----------------------------------------------------------------------------------------
# PRE-CONDICIONES
# ----------------------------------------------------------------------------------------
# - El valor de 'a' debe ser diferente de 0
# - Los valores ingresados deben ser números enteros

# ----------------------------------------------------------------------------------------
# POSTCONDICIONES
# ----------------------------------------------------------------------------------------
# - El programa muestra las soluciones X1 y X2 de la ecuación cuadrática
# - Si no existen soluciones reales, se informa al usuario

# ----------------------------------------------------------------------------------------
# PARAMETROS
# ----------------------------------------------------------------------------------------
# a: coeficiente cuadrático
# b: coeficiente lineal
# c: término independiente

# ----------------------------------------------------------------------------------------
# CUERPO DEL PROGRAMA
# ----------------------------------------------------------------------------------------

# Entrada de datos
a = int(input("Ingrese el valor de a: "))
b = int(input("Ingrese el valor de b: "))
c = int(input("Ingrese el valor de c: "))

# Cálculo del discriminante
discriminante = b**2 - 4*a*c

# Proceso y salida
if discriminante < 0:
    print("La ecuación no tiene soluciones reales.")
else:
    x1 = (-b + math.sqrt(discriminante)) / (2*a)
    x2 = (-b - math.sqrt(discriminante)) / (2*a)

    print("La solución X1 es:", x1)
    print("La solución X2 es:", x2)
    if discriminante == 0:
      print ('Solamente se cuenta con una raiz x1=x2')

# ----------------------------------------------------------------------------------------
# end.
# ----------------------------------------------------------------------------------------

# Ejemplos para hacer la prueba:
# a= 3; b= -6; c= 3 ----- Una solución (osea la parábola toca una sola vez el suelo)
# a= 1; b= 2; c= -4 -----  Dos soluciones
# a= 1; b= 2; c= 4 ---- no hay solucion en los números reales.
```
