```python
# ----------------------------------------------------------------------------------------
# PROGRAMA: Orugas
# ----------------------------------------------------------------------------------------
# Descripcion: Simula la supervivencia de dos orugas (amarilla y verde) que avanzan
# por una rama de 1 m hacia un fruto. Cada oruga tiene un patron de avance distinto
# y la rama crece 10 cm cada noche. Las orugas solo resisten 10 dias sin comer.
# ----------------------------------------------------------------------------------------
# Autor: Grupo 10 (Leidy Villamil, Juan P. Angulo, Sergio Camargo, Julian Soto, Juan Ayala)
# Version: 1.0
# [12.02.2025]
# ----------------------------------------------------------------------------------------

# ----------------------------------------------------------------------------------------
# PRE-CONDICIONES
# ----------------------------------------------------------------------------------------
# - La rama mide inicialmente 100 cm.
# - Las orugas tienen un maximo de 10 dias para llegar al fruto.
# - Oruga amarilla: avanza 1 cm el primer dia, y cada dia el doble del anterior.
# - Oruga verde: avanza 2 cm los dos primeros dias, y luego la suma de los dos dias anteriores.
# - La rama crece 10 cm cada noche.

# ----------------------------------------------------------------------------------------
# POSTCONDICIONES
# ----------------------------------------------------------------------------------------
# - El programa muestra dia a dia la longitud de la rama, el avance y la posicion acumulada.
# - Indica si la oruga llego al fruto o murio en el intento.

# ----------------------------------------------------------------------------------------
# PARAMETROS
# ----------------------------------------------------------------------------------------
# rama: longitud actual de la rama en cm (inicia en 100)
# posicion: distancia total recorrida por la oruga en cm
# avance: distancia que recorre la oruga en el dia actual
# dia: numero del dia actual (de 1 a 10)
# avance_anterior: (solo oruga verde) avance del dia anterior
# avance_nuevo: (solo oruga verde) calculo del proximo avance

# ----------------------------------------------------------------------------------------
# CODIGO
# ----------------------------------------------------------------------------------------

# ========== ORUGA AMARILLA ==========
print("-.-.-.-.-.-. ORUGA AMARILLA -.-.-.-.-..-")

dia = 1
posicion = 0
rama = 100
avance = 1

while dia <= 10 and posicion < rama:
    posicion = posicion + avance

    print("Dia", dia, "---> Posicion:",
         posicion, "cm ---> Rama:",
             rama, "cm ---> Avance:", avance, "cm")

    if posicion < rama:
        rama = rama + 10
        avance = avance * 2
        dia = dia + 1

if posicion >= rama:
    print("La oruga amarilla llego al fruto el dia", dia)
else:
    print("La oruga amarilla murio sin llegar")

# ========== ORUGA VERDE ==========
print()
print(".-.-.-.-.-.-- ORUGA VERDE .-.-.-.-.-.")

rama = 100
posicion = 0
avance = 2
avance_anterior = 0
dia = 1

while dia <= 10 and posicion < rama:
    posicion = posicion + avance

    print("Dia", dia, "---> Posicion:",
         posicion, "cm ---> Rama:",
             rama, "cm ---> Avance:", avance, "cm")

    if posicion < rama:
        rama = rama + 10

        if dia < 2:
            avance_nuevo = 2
        else:
            avance_nuevo = avance + avance_anterior

        avance_anterior = avance
        avance = avance_nuevo
        dia = dia + 1

if posicion >= rama:
    print("La oruga verde llego al fruto el dia", dia)
else:
    print("La oruga verde murio sin llegar")

# ----------------------------------------------------------------------------------------
# end.
# Se usa un bucle while para simular dia a dia el avance de cada oruga.
# La oruga amarilla duplica su avance cada dia (1, 2, 4, 8...).
# La oruga verde sigue un patron Fibonacci (2, 2, 4, 6, 10, 16...).
# Ambas orugas logran llegar al fruto antes de los 10 dias.
# ----------------------------------------------------------------------------------------
```
