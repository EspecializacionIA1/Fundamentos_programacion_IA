```python

%%writefile G10_funcionesPandas.py
# ----------------------------------------------------------------------------------------
# PROGRAMA: G10_funcionesPandas.py
# ----------------------------------------------------------------------------------------
# Descripción: Módulo con las funciones de exploración de datos.
#              Contiene las funciones que serán importadas por G10_UsoPandas.py
#              para analizar la tabla de posiciones de la Liga BetPlay 2025.
# ----------------------------------------------------------------------------------------
# Autor: Julian José Soto Gómez, Juan Gilberto Ayala Salazar,
# Version: 1.0
# [05.03.2026]
# ----------------------------------------------------------------------------------------
# IMPORTAR MODULOS
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# ----------------------------------------------------------------------------------------
# PRE-CONDICIONES
# ----------------------------------------------------------------------------------------

# - El módulo pandas debe estar instalado en el entorno de ejecución.
# - El módulo matplotlib debe estar instalado en el entorno de ejecución.
# - Todas las funciones reciben como parámetro un DataFrame de Pandas previamente cargado.
# - El DataFrame debe contener las columnas: EQUIPO, PT, PJ, PG, PE, PP, GF, GC, GFV, DG.
# - Los valores de las columnas numéricas no deben ser nulos ni estar vacíos.
# - Para configurar_series: el DataFrame debe tener al menos 9 filas (fila 0 = Medellín, fila 8 = Junior).
# - Para graficar_puntos: el directorio de ejecución debe tener permisos de escritura
#   para guardar el archivo puntos_liga2025.png.

# ----------------------------------------------------------------------------------------
# POSTCONDICIONES
# ----------------------------------------------------------------------------------------

# - Las funciones no retornan valores; su resultado es la impresión en consola.
# - resumen_general: imprime todos los registros y estadísticas descriptivas del DataFrame.
# - datos_destacados: imprime equipos destacados, promedio de puntos y total de goles.
# - ranking_rendimiento: agrega la columna "% Victorias" al DataFrame e imprime el ranking.
# - graficar_puntos: genera y guarda el archivo "puntos_liga2025.png" en el directorio actual.
# - configurar_series: imprime 4 Series extraídas del DataFrame (2 por columna, 2 por fila).

# ----------------------------------------------------------------------------------------
# PARAMETROS
# ----------------------------------------------------------------------------------------
# - dataframe: DataFrame de Pandas con la tabla de posiciones de la Liga BetPlay 2025.
#              Debe contener las columnas: EQUIPO, PT, PJ, PG, PE, PP, GF, GC, GFV, DG.
# ----------------------------------------------------------------------------------------

def resumen_general(dataframe):
    print("\n***** FUNCIÓN 1: Resumen general del DataFrame *****")
    print("\nTodos los registros:")
    print(dataframe.to_string(index=False))
    print("\nEstadísticas básicas de las columnas numéricas:")
    print(dataframe.describe())


def datos_destacados(dataframe):
    print("\n***** FUNCIÓN 2: Datos destacados *****")
    equipo_lider  = dataframe.loc[dataframe["PT"].idxmax(), "EQUIPO"]
    equipo_ultimo = dataframe.loc[dataframe["PT"].idxmin(), "EQUIPO"]
    mas_goles     = dataframe.loc[dataframe["GF"].idxmax(), "EQUIPO"]
    menos_goles   = dataframe.loc[dataframe["GC"].idxmin(), "EQUIPO"]
    print(f"\nEquipo líder (más puntos)   : {equipo_lider}")
    print(f"Equipo último (menos puntos): {equipo_ultimo}")
    print(f"Equipo más goleador         : {mas_goles}")
    print(f"Equipo menos goleado        : {menos_goles}")
    print(f"\nPromedio de puntos en la liga: {dataframe['PT'].mean():.1f}")
    print(f"Total de goles en la liga   : {dataframe['GF'].sum()}")


def ranking_rendimiento(dataframe):
    print("\n***** FUNCIÓN 3: Ranking de rendimiento por equipo *****")
    dataframe["% Victorias"] = ((dataframe["PG"] / dataframe["PJ"]) * 100).round(1)
    ranking = dataframe[["EQUIPO", "PJ", "PG", "PE", "PP", "PT", "% Victorias"]]
    ranking = ranking.sort_values("% Victorias", ascending=False)
    print("\nEquipos ordenados por porcentaje de victorias:\n")
    print(ranking.to_string(index=False))
    print(f"\nEquipo con mejor rendimiento : {ranking.iloc[0]['EQUIPO']} ({ranking.iloc[0]['% Victorias']}%)")
    print(f"Equipo con peor rendimiento  : {ranking.iloc[-1]['EQUIPO']} ({ranking.iloc[-1]['% Victorias']}%)")


def graficar_puntos(dataframe):
    print("\n--- FUNCIÓN 4: Gráfica de puntos por equipo ---")
    dataframe.plot(kind="bar", x="EQUIPO", y="PT",
                   title="Puntos por equipo — Liga BetPlay 2025",
                   xlabel="Equipo", ylabel="Puntos",
                   color="steelblue", legend=False, figsize=(12, 5))
    plt.xticks(rotation=45, ha="right", fontsize=7)
    plt.tight_layout()
    plt.savefig("puntos_liga2025.png", dpi=150)
    print("Gráfica guardada como: puntos_liga2025.png")

def configurar_series(dataframe):
    print("\n***** SERIES: Configuración de índices *****")

    # Establecemos la columna EQUIPO como índice del DataFrame
    # para que las Series queden etiquetadas con el nombre del equipo
    df_idx = dataframe.set_index("EQUIPO")

    # --- Serie 1 (desde columna): Puntos de cada equipo ---
    serie_puntos = df_idx["PT"]
    print("\nSerie 1 — Puntos (PT) por equipo:")
    print(serie_puntos.to_string())

    # --- Serie 2 (desde columna): Diferencia de goles de cada equipo ---
    serie_dg = df_idx["DG"]
    print("\nSerie 2 — Diferencia de goles (DG) por equipo:")
    print(serie_dg.to_string())

    # --- Serie 3 (desde fila): Todos los datos de Independiente Medellín ---
    serie_medellin = dataframe.loc[0]
    print("\nSerie 3 — Datos de Independiente Medellín (fila 0):")
    print(serie_medellin.to_string())

    # --- Serie 4 (desde fila): Todos los datos de Junior ---
    serie_junior = dataframe.loc[8]
    print("\nSerie 4 — Datos de Junior (fila 8):")
    print(serie_junior.to_string())


# ----------------------------------------------------------------------------------------
# end.
# ----------------------------------------------------------------------------------------
```

```python
%%writefile G10_UsoPandas.py
# ----------------------------------------------------------------------------------------
# PROGRAMA: G10_UsoPandas.py
# ----------------------------------------------------------------------------------------
# Descripción: Carga el archivo Excel con la tabla de posiciones de la Liga
#              BetPlay 2025 y llama las funciones de exploración definidas
#              en el módulo G10_funcionesPandas.py
# ----------------------------------------------------------------------------------------
# Autor: Julian José Soto Gómez, Juan Gilberto Ayala,
# Version: 1.0
# [05.03.2026]
# ----------------------------------------------------------------------------------------
# IMPORTAR MODULOS
import pandas as pd
from G10_funcionesPandas import resumen_general, datos_destacados, ranking_rendimiento, graficar_puntos, configurar_series

# ----------------------------------------------------------------------------------------
# PRE-CONDICIONES
# ----------------------------------------------------------------------------------------

# El archivo TablaPosiciones2025.xlsx debe estar en la misma carpeta que este programa.
# El módulo G10_funcionesPandas.py también debe estar en la misma carpeta.

# ----------------------------------------------------------------------------------------
# POSTCONDICIONES
# ----------------------------------------------------------------------------------------

# - Se carga el archivo CSV/Excel y se construye el DataFrame df correctamente.
# - Se invocan las funciones del módulo G10_funcionesPandas.py en el siguiente orden:
#   configurar_series → resumen_general → datos_destacados → ranking_rendimiento → graficar_puntos
# - Todos los resultados se imprimen en consola y la gráfica queda guardada como PNG.

# ----------------------------------------------------------------------------------------
# PARAMETROS
# ----------------------------------------------------------------------------------------
# - df: DataFrame cargado desde TablaPosiciones2025.xlsx
#       Contiene 20 filas (equipos) y 11 columnas con los datos de la Liga BetPlay 2025.

# ----------------------------------------------------------------------------------------

df = pd.read_excel("TablaPosiciones2025.xlsx")

print("===============================================================================")
print("  PUNTO 2 — CONFIGURACIÓN DE ÍNDICES – SERIES")
print("===============================================================================")

configurar_series(df)

print("===============================================================================")
print("  PUNTO 3 — EXPLORACIÓN DE DATOS")
print("===============================================================================")

resumen_general(df)
datos_destacados(df)
ranking_rendimiento(df)
graficar_puntos(df)


# ----------------------------------------------------------------------------------------
# end.
# ----------------------------------------------------------------------------------------
```
