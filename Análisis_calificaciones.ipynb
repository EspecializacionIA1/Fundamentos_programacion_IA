```python
%%writefile funciones.py

def showMenu() -> int:
    print("\n========================================")
    print(" MENÚ CALIFICACIONES DE ESTUDIANTES")
    print("\n========================================")
    print("1. Registrar un nuevo grupo")
    print("2. Ordenar la lista")
    print("3. Buscar por calificación")
    print("4. Salir")
    option = int(input("  Ingrese la opción deseada (1-4) o 0 para terminar: "))
    return option

def registerNewGroup():
  studentsList = []

 # Ingreso numero de estudiantes
  while True:
    n = int(input("Ingrese el número de estudiantes del grupo "))
    if n <= 0:
      print("Debe ingresar un número correcto de estudiantes")
    else:
      break

   #Ingresar apellido de los estudiantes
  for i in range(n):
    print(f"\n Información del estudiante {i + 1} de {n}")

    while True:
      lastName = input(f"   Ingrese el apellido del estudiante {i + 1} :")

      #Validar que el apellido del estudiante no esté repetido dentro de la lista
      if lastName == "":
        print("El apellido del estudiante no puede ser vacío. Ingrese un apellido válido")
      elif any(c.isdigit() for c in lastName):
        print("No puede ingresar números para un apellido. Ingrese un apellido con solo letras")
      elif any(est[0].lower() == lastName.lower() for est in studentsList):
        print(f"El apellido '{lastName}' ya está registrado. Use uno diferente.")
      else:
          break



    while True:
      score = float(input(f"ingrese la calificación (0.0 - 5.0) para el estudiante: {i + 1} "))

    #lógica para validación de calificaciones
      if( 0.0 <= score <= 5.0):
        break
      else:
        print("La calificación ingresada es incorrecta. Ingrese solo números decimales (use punto decimal)")

    studentsList.append((lastName, score))
    #:.1f --> es para redondear el valor decimal.
    print(f" Estudiante: {lastName} registrado correctamente con la nota: {score:.1f}")

  print(f"El grupo fué registrado exitosamente con {n} estudiantes")
  return studentsList
```

```python
%%writefile principal.py

from funciones import (showMenu, registerNewGroup)

option = showMenu()

while option != 0:
  if option == 1:
    registerNewGroup()
  elif option == 4:
    print("Hasta luego!")
    break
  else:
    print("La opción ingresada no es válida")

  option = showMenu()





```
