````python
"""

1. Gráfica 1: Ecuación 6: teniendo en cuenta que cos(a) corresponde al lado abyacente=posición horizontal, y sen(a) corrreponde al lado opuesto=posición vertical, al ir cambiando el ángulo:

t * cos(t) = oscila ed lado a lago.
t = tiempo, entonces avanza en línea recta.
t * sen(t) = oscila de arrija y abajo.

Entonces al usar cos y sen mutiplciados por t, esto hace que un punto gire en círculo y esto multiplicado por t=tiempo, nos da una espiral que se va expandiento.

"""

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

t = np.linspace(0, 2 * np.pi, 1000)

x = t * np.cos(t)
y = t
z = t * np.sin(t)

ax.plot(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('(t·cos(t), t, t·sin(t))')

plt.tight_layout()
````

````python

"""

Gráfica 2: Ecuación 3.
cos(t), sin(t) como se explicó en la gráfica anterior, hay un giro en forma de círculo.
y tenemos por otro lado una división de 1 por lo que es la representación de un número en el tiempo cada vez mas pequeño pero con mas dígitos.
como el denominador es cada vez mas grande (en digitos), el resultado es cada vez mas pequeño: p.e. 1/2=0.5, 1/20=0.1, 1/100=0.01 ...
como vemos en la gráfica, nunca toca el piso, osea nunca es igual a 0.
"""


import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

t = np.linspace(0, 2 * np.pi, 1000)

x = np.cos(t)
y = np.sin(t)
z = 1 / (t**2 + 1)

ax.plot(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('(cos(t), sin(t), 1/(t²+1))')

plt.tight_layout()
````

````python
"""
Gráfica 3: Ecuación 1:
tenemos z=t**2. No es un movimiento simple como t=tiempo que aumenta en un movimiento "parejo",
sino que al multiplicar t * t**2 esto hace un momento cada vez más rápido como una parábola o una rampa.
Al mismo tiempo se va "pegando a la pared" ya que como explicabamos en grafica anterior 1/(t²+1) su denominador es cada vez mas grande y baja de 1 hacia cero.(pero nunca cero)
Es por este movimiento que concluimos que la ecuación 1 corresponde a la gráfica 3.
"""

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

t = np.linspace(0, 2 * np.pi, 1000)

x = t
y = 1 / (t**2 + 1)
z = t**2

ax.plot(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('(t, 1/(t²+1), t²)')

plt.tight_layout()
````

````python

"""
Gráfica 5: ecuación 4.

Teniendo en cuenta que cos(a) corresponde al lado abyacente=posición horizontal, y sen(a) corrreponde al lado opuesto=posición vertical,
esto genera un movimiento circular cuando hacemos la multplicación por t.
como estamos multiplicando 8 * t(tiempo), esto generará un movimiento mmucho mas rápido.
Tambien partimos de la defición de e(euler) que aparece en la ecucación, sabiendo que es un valor que crece exponencialmente,
podemos concluir que se hace un movimiento circular de manera rápida ascendente de manera exponencial. Gráfica 5.
"""


import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

t = np.linspace(0, 2 * np.pi, 1000)

x = np.cos(8 * t)
y = np.sin(8 * t)
z = np.exp(0.8 * t)

ax.plot(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('(cos(8t), sin(8t), e^(0.8t))')

plt.tight_layout()
````

````python

"""
Gráfica 6: Ecuación 2: Aplicando el teorema de pitágoras: a**2 + b**2 = c**2 (suma de cuadrados de los catetos, igual a cuadrado de hipotenusa),
llegamos a la identidad fundamental: cos²(t) + sin²(t) = 1.
Entonces como cos y sen están elevados al cuadrado (cualquier número elevado al cuadrado nunca será negativo),
esto quiere decir que cuando cos sea = 0, sen será igual a 1 y viceversa mientras que t=tiempo subirá constantemente,
generando de esta manera un movimiento en zigzag como lo podemso ver en la gráfica 6.
"""

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

t = np.linspace(0, 2 * np.pi, 1000)

x = np.cos(t)**2
y = np.sin(t)**2
z = t

ax.plot(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('(cos²(t), sin²(t), t)')

plt.tight_layout()
````

##Ejercicio 10

considere la función lógisitca
$$f(x)=\dfrac{1}{1+e^{x}}$$
Si $g(x,y,a,b)=f(-2f(3x+5y-a)+3f(-2x+4y+b))$, use la regla de la cadena para hallar $\frac{\partial f}{\partial x}$ y $\frac{\partial f}{\partial a}$ su derivada

<h2> Solucion 10 </h2>

````python
def f(x):
    return 1 / (1 + np.exp(x))

def f_prima(x):
    return -f(x) * (1 - f(x))

def calcular_todo(x, y, a, b):
    u = 3*x + 5*y - a
    v = -2*x + 4*y + b
    w = -2*f(u) + 3*f(v)

    dg_dx = -6 * f_prima(w) * (f_prima(u) + f_prima(v))
    dg_da = 2 * f_prima(w) * f_prima(u)

    return dg_dx, dg_da

# Probamos con diferentes valores
print("Resultados para diferentes valores:")
print("=" * 65)
print(f"{'x':>3} {'y':>3} {'a':>3} {'b':>3} │ {'∂g/∂x':>15} │ {'∂g/∂a':>15}")
print("─" * 65)

for x, y, a, b in [(1,2,3,4), (0,0,0,0), (1,1,1,1), (2,3,1,5)]:
    dg_dx, dg_da = calcular_todo(x, y, a, b)
    print(f"{x:>3} {y:>3} {a:>3} {b:>3} │ {dg_dx:>15.8f} │ {dg_da:>15.8f}")
````
